```python
import scipy.stats as ss
```

```python
n = 20
s = 1/5       #Probability of giving the right answer
p = 1-s       # Probability of giving wrong answer
max_p = 5     #max number of wrong answers

hh = ss.binom(n,p)

total_p = hh.pmf(max_p)
```

```python
print('{:.7f}'.format(total_p))       #This will remove the scientific notation and will print the answer up to 7 decimal places.
```
